```
# SET PATHS
import os 
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"
os.environ["SPARK_HOME"] = "/content/spark-2.4.0-bin-hadoop2.7"

# INITIALIZE FINDSPARK
import findspark
findspark.init()

# CREATE CONTEXT
import pyspark
try:
  sc
except NameError:
  sc = pyspark.SparkContext(appName="xyz")
  
```

```
lines = sc.textFile("big.txt")
```

```
lines.count()
```

```
lines.take(5)
```

```
import string

def getWords(line):
   words = line.strip().lower().split(' ')
   return words

words = lines.flatMap(getWords).map(lambda x: (x.strip(string.punctuation), 1))
```

```
words.count()
```

```
words.getNumPartitions()
```

```
words.take(5)
```

```
wordcount = words.reduceByKey(lambda x,y: x+y)
```

```
wordcount.take(5)
```

```
countword = wordcount.map(lambda p: (p[1],p[0]))
```

```
countword.take(5)
```

```
countword.sortByKey(False).take(50)
```
